## Basic Image Enhancements

### Addition
Memakai OpenCV, bukalah file ikon kotamu, lalu:
1. Di tiap pixel, di tiap channel (BRG), kurangi nilainya sebesar 60
2. Di tiap pixel, di tiap channel (BRG), tambahi nilainya sebesar 60

Tampilkan hasilnya dan beri komentar, apa bedanya dengan gambar aslinya.


```python
## Ikon kota
tbn_rgb = cv2.cvtColor(image_tbn, cv2.COLOR_BGR2RGB)

matrix = np.ones(tbn_rgb.shape, dtype = "uint8") * 60

tbn_rgb_cerah   = cv2.add(tbn_rgb, matrix)
tbn_rgb_gelap   = cv2.subtract(tbn_rgb, matrix)

# Tampilkan gambar
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(tbn_rgb_gelap); plt.title("Gelap");
plt.subplot(132); plt.imshow(tbn_rgb);       plt.title("Original");
plt.subplot(133); plt.imshow(tbn_rgb_cerah); plt.title("Terang");
```

##### Komentar
Gambar yang pixel nya ditambah sebesar 60 menggunakan fungsi add menjadi lebih terang dan gambar yang pixel nya dikurangi sebesar 60 menggunakan fungsi subtract menjadi lebih gelap 

### Multiplication
Memakai OpenCV, bukalah file gambar ikon kotamu, lalu:
1. Di tiap pixel, di tiap channel (BRG), kalikan nilainya sebesar 0.7
2. Di tiap pixel, di tiap channel (BRG), tambahi nilainya sebesar 1.3

Tampilkan hasilnya dan beri komentar, apa bedanya dengan gambar aslinya.


```python
## Ikon kota
matrix1 = np.ones(tbn_rgb.shape) * .7
matrix2 = np.ones(tbn_rgb.shape) * 1.3

tbn_rgb_gelap   = np.uint8(cv2.multiply(np.float64(tbn_rgb), matrix1))
tbn_rgb_terang  = np.uint8(cv2.multiply(np.float64(tbn_rgb), matrix2))

# Tampilkan Gambar
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(tbn_rgb_gelap);  plt.title("Kontras Rendah");
plt.subplot(132); plt.imshow(tbn_rgb);        plt.title("Original");
plt.subplot(133); plt.imshow(tbn_rgb_terang); plt.title("Kontras Tinggi");
```

##### Komentar
Gambar dengan kontras rendah akan terlihat sedikit lebih gelap dan gambar dengan kontras tinggi terlihat lebih terang tetapi terjadi overflow pada warna-warna yang terang di kontras tinggi

#### Jika ada overflow, cegah dengan clip
Tampilkan hasilnya dan beri komentar, apa bedanya dengan gambar aslinya.

```python
## Mencegah overflow dengan clip
tbn_rgb_terang  = np.uint8(np.clip(cv2.multiply(np.float64(tbn_rgb), matrix2),0,255))

# Tampilkan Gambar
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(tbn_rgb_gelap);  plt.title("Kontras Rendah");
plt.subplot(132); plt.imshow(tbn_rgb);        plt.title("Original");
plt.subplot(133); plt.imshow(tbn_rgb_terang); plt.title("Kontras Tinggi");
```

##### Komentar
Sama seperti hasil sebelumnya, setelah menggunakan fungsi clip gambar dengan kontras tinggi tidak terjadi overflow

### Image Thresholding
Memakai OpenCV, 
1. Bukalah file gambar ikon kotamu sebagai grayscale image
2. Ubah gambar grayscale itu menjadi gambar biner, dengan thresholding 50, 127, 200

Tampilkan hasilnya dan beri komentar. Apa beda di antara ketiganya?


```python
## Thresholding
tbn_read = cv2.imread("tuban_kuy.jpg", cv2.IMREAD_GRAYSCALE)

# thresholding 1
retval, tbn_thresh_1 = cv2.threshold(tbn_read,50, 255, cv2.THRESH_BINARY)

# thresholding 2
retval, tbn_thresh_2 = cv2.threshold(tbn_read,127, 255, cv2.THRESH_BINARY)

# thresholding 3
retval, tbn_thresh_3 = cv2.threshold(tbn_read,200, 255, cv2.THRESH_BINARY)

# Tampilkan Gambar
plt.figure(figsize=[18,15])
plt.subplot(221); plt.imshow(tbn_read,     cmap="gray");  plt.title("Original");
plt.subplot(222); plt.imshow(tbn_thresh_1, cmap="gray");  plt.title("Thresholded (global: 50)");
plt.subplot(223); plt.imshow(tbn_thresh_2, cmap="gray");  plt.title("Thresholded (global: 127)");
plt.subplot(224); plt.imshow(tbn_thresh_3, cmap="gray");  plt.title("Thresholded (global: 200)");
```

##### Komentar
Gambar dengan nilai thresold rendah dominan warna yang ditampilkan putih sedangkan dengan nilai threshold tinggi dominan akan menampilkan warna hitam

### Application
Memakai `ums-line.png` dan `SW_1280x720.jpg`, buatlah tulisan 'UMS' menjadi foreground, dan gambar gedung SW sebagai background. Tulisan putih 'UMS' akan diganti dengan gambar gedung. Lakukan seperti yang ada di video.

Jika kamu akan me-resize gambar gedungnya, aspect ratio -nya tidak boleh berubah. Kamu juga dipersilakan meng-crop (memakai OpenCV) gambar gedung agar ukurannya sama dengan foreground.

```python
## Menaruh gambar gedung di belakang.
# im_foreground = 
img_bgr = cv2.imread("ums-line.png")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=[5,5])
logo_w = img_rgb.shape[0]
logo_h = img_rgb.shape[1]
plt.imshow(img_rgb)
print(img_rgb.shape)
```

```python
# im_background = 
img_background_bgr = cv2.imread("SW_1280x720.jpg")
img_background_rgb = cv2.cvtColor(img_background_bgr, cv2.COLOR_BGR2RGB)

potong = img_background_rgb[100:551, 100:956]

plt.figure(figsize=[5,5])
plt.imshow(potong) 
print(potong.shape)
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# Apply global thresholding to creat a binary mask of the logo
retval, img_mask = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)

# Create an inverse mask
img_mask_inv = cv2.bitwise_not(img_mask)
plt.figure(figsize=[5,5])
plt.imshow(img_mask_inv,cmap="gray")
```

```python
# Create colorful background "behind" the logo lettering
img_background = cv2.bitwise_and(potong, potong, mask=img_mask)
plt.figure(figsize=[5,5])
plt.imshow(img_background)
```

```python
# Isolate foreground (red from original image) using the inverse mask
img_foreground = cv2.bitwise_and(img_rgb, img_rgb, mask=img_mask_inv)
plt.figure(figsize=[5,5])
plt.imshow(img_foreground)
```

```python
# Add the two previous results obtain the final result
result = cv2.add(img_background,img_foreground)
plt.figure(figsize=[5,5])
plt.imshow(result)
```
