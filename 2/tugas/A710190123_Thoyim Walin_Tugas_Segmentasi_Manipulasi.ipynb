## Color segmentation

### Memakai OpenCV, extractlah bunga warna kuning pada file yang sudah disertakan

```python
## Buka Florist_Shop.jpg
#  Convert ke HSV
#  Tentukan nilai H,S,V untuk warna yang akan diextract.
#  Extract bunga warna kuning. Tampilkan mask-nya, hasil extract, dan gambar asli.
# ...
```

```python
im_bunga = cv2.imread('Florist_Shop.jpg')
im_bunga.shape
```

```python
plt.imshow(im_bunga[:,:,::-1])
```

```python
hsv_bunga = cv2.cvtColor(im_bunga,cv2.COLOR_BGR2HSV)
```

```python
hsv_bunga.shape
```

```python
kuning_bawah = np.array([25,180,0])
kuning_atas  = np.array([35,255,255])
mask2 = cv2.inRange(hsv_bunga,kuning_bawah,kuning_atas)
```

```python
plt.imshow(mask2, cmap='gray')
```

```python
res2 = cv2.bitwise_and(im_bunga,im_bunga,mask=mask2)
```

```python
plt.imshow(res2[:,:,::-1])
```

```python
plt.figure(figsize=[20,10])
plt.subplot(141);plt.imshow(mask2, cmap='gray');plt.title("Hasil Mask");
plt.subplot(142);plt.imshow(res2[:,:,::-1]);plt.title("Hasil Extract");
plt.subplot(143);plt.imshow(im_bunga[:,:,::-1]);plt.title("Gambar Asli");
```

## Resizing

Lakukan pengubahan ukuran gambar seperti yang di video. Pakailah hasil crop ikon kotamu sebagai gambar yang diubah ukurannya.

```python
## Metode 1
# ...
patung_2x = cv2.resize(patung,None,fx=2, fy=2)
plt.imshow(patung_2x)
```

```python
## Metode 2
# ...
desired_width = 200
desired_height = 100
dim = (desired_width, desired_height)

# Resize background image to sae size as logo image
resized_patung = cv2.resize(patung, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_patung)
```

```python
## Mempertahankan aspect ratio:
# ...
desired_width = 100
aspect_ratio = desired_width / patung.shape[1]
desired_height = int(patung.shape[0] * aspect_ratio)
dim = (desired_width, desired_height)

# Resize image
resized_patung = cv2.resize(patung, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_patung)
```

## Flipping Images
Memakai gambar ikon kotamu yang utuh, lakukan perputaran seperti yang di video. Tampilkan semuanya, termasuk gambar yang asli.

```python
img_tuban_bgr = cv2.imread("tuban_kuy.jpg",cv2.IMREAD_COLOR)
img_tuban_rgb = img_tuban_bgr[:,:,::-1] 

img_tuban_rgb_flipped_horz = cv2.flip(img_tuban_rgb, 1)
img_tuban_rgb_flipped_vert = cv2.flip(img_tuban_rgb, 0)
img_tuban_rgb_flipped_both = cv2.flip(img_tuban_rgb, -1)

# Show the images
plt.figure(figsize=[18,5])
plt.subplot(141);plt.imshow(img_tuban_rgb_flipped_horz);plt.title("Horizontal Flip");
plt.subplot(142);plt.imshow(img_tuban_rgb_flipped_vert);plt.title("Vertical Flip");
plt.subplot(143);plt.imshow(img_tuban_rgb_flipped_both);plt.title("Both Flipped");
plt.subplot(144);plt.imshow(img_tuban_rgb);plt.title("Original");
```
